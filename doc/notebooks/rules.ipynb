## Copy rule

```python
with open('../diagrams/copy.json', 'r') as file: js = json.load(file)

g = Graph.from_json(js)
zx.draw(g)
```

```python
zx.simplify.to_gh(g)
zx.simplify.simp(g, 'copy_simp', rules.match_copy, rules.apply_copy)

zx.draw(g)
```

## Spider Fusion

```python
with open('../diagrams/spiderFuse.json', 'r') as file: js = json.load(file)

g = Graph.from_json(js)
zx.draw(g)
```

```python
zx.simplify.simp(g, 'spider_simp', rules.match_spider_parallel, rules.spider)
zx.draw(g)
```
